```python
import numpy as np
import sys
import matplotlib.pyplot as plt
import math
plt.style.use('seaborn-poster')
```

```python
NUMERO_TERMOS = 20
```

```python
def arctan_taylor(x, n):
    return np.power(-1,n) * ((x ** ((2 * n) + 1)) / (2 * n + 1))

def serie_arctan_taylor(x, n):
    result = [arctan_taylor(x, j) for j in range(n)]
    return np.array(result)


def calc_erro_rel(serie):
    erros_aux = []
    for j in range(1, serie.size):
        erro = np.abs((serie[j-1] - serie[j])/serie[j-1])
        erros_aux.append(erro)
    return np.array(erros_aux)

def calc_erro_real(serie):
    erros_aux = []
    for j in range(NUMERO_TERMOS):
        erros_aux.append(np.abs(1-(serie[j]/np.pi)))
    return np.array(erros_aux)
```

```python
def plot_grafico_erro(erros, label):
    y_axis = erros
    x_axis = range(NUMERO_TERMOS-1)
    plt.figure(figsize = (10,8))
    plt.figure()
    plt.title(f'Gráfico de erro para cálculo de pi pela formula de {label}')
    plt.plot(x_axis, y_axis, 'k')
    plt.grid()
    plt.xlabel('$x$')
    plt.ylabel('$Erro$')
```

## Formula de Machin

```python
res = 0
x1 = 1/5
x2 = 1/239
resultadosMachinAux = []
for i in range(NUMERO_TERMOS):
    res = 4*(4 * serie_arctan_taylor(x1, i) - serie_arctan_taylor(x2, i))
    resultadosMachinAux.append(res.sum())
    print(f'Usando {i} termo: {res.sum()}')
resultadosMachin = np.array(resultadosMachinAux)
```

```python
plot_grafico_erro(calc_erro_rel(resultadosMachin), 'Machin')
```

## Formula de Hutton

```python
res = 0
x1 = 1/2
x2 = 1/3
resultadosHuttonAux = []
for i in range(NUMERO_TERMOS):
    res = 4*(serie_arctan_taylor(x1, i) + serie_arctan_taylor(x2, i))
    resultadosHuttonAux.append(res.sum())
    print(f'Usando {i} termo: {res.sum()}')
resultadosHutton = np.array(resultadosHuttonAux)
```

```python
plot_grafico_erro(calc_erro_rel(resultadosHutton), 'Hutton')
```

## Fórmula de Clausen

```python
res = 0
x1 = 1/3
x2 = 1/7
resultadosClausenAux = []
for i in range(NUMERO_TERMOS):
    res = 4*(2*serie_arctan_taylor(x1, i) + serie_arctan_taylor(x2, i))
    resultadosClausenAux.append(res.sum())
    print(f'Usando {i} termo: {res.sum()}')
resultadosClausen = np.array(resultadosClausenAux)
```

```python
plot_grafico_erro(calc_erro_rel(resultadosClausen), 'Clausen')
```

## Fórmula de Dase

```python
res = 0
x1 = 1/2
x2 = 1/5
x3 = 1/8
resultadosDaseAux = []
for i in range(NUMERO_TERMOS):
    res = 4*(serie_arctan_taylor(x1, i) + serie_arctan_taylor(x2, i)+serie_arctan_taylor(x3, i))
    resultadosDaseAux.append(res.sum())
    print(f'Usando {i} termo: {res.sum()}')
resultadosDase = np.array(resultadosDaseAux)
```

```python
plot_grafico_erro(calc_erro_rel(resultadosDase), 'Dase')
```

## Calculando os erros comparando com o valor de np.pi

```python
erroRealMachin = calc_erro_real(resultadosMachin)
erroRealHutton = calc_erro_real(resultadosHuttonAux)
erroRealDase = calc_erro_real(resultadosDase)
erroRealClausen = calc_erro_real(resultadosClausen)
```

```python
plt.figure(figsize = (10,8))
plt.plot(np.arange(0,NUMERO_TERMOS),erroRealClausen,label='Machin')
plt.plot(np.arange(0,NUMERO_TERMOS),erroRealClausen,label='Clausen')
plt.plot(np.arange(0,NUMERO_TERMOS),erroRealDase,label='Dase')
plt.plot(np.arange(0,NUMERO_TERMOS),erroRealHutton,label='Hutton')

plt.legend()
```
